## Part 4: Import JavaScript Object Notation .json files

<h4>JSON to Pandas DataFrame Using json_normalize()</h4>
<p>
Pandas and JSON libraries in Python can help in achieving this. We have two functions read_json() and json_normalize() which can help in converting JSON string to a DataFrame.</p>

```python
import json
from pandas import json_normalize

data = '''
{
"Results":
         [
         { "id": "1", "Name": "Jay" },
         { "id": "2", "Name": "Mark" },
         { "id": "3", "Name": "Jack" }
         ],
"status": ["ok"]
}
    '''

info = json.loads(data)

df6 = json_normalize(info['Results']) #Results contain the required data
```

```python
df6.head()
```

```python
df6.info()
```

<h4>JSON to Pandas DataFrame Using read_json()</h4>
<p> We can directly pass the path of a JSON file or the JSON string to the function for storing data in a Pandas DataFrame. read_json() has many parameters, among which orient specifies the format of the JSON string.</p>

```python
data = '''
          {
              "0":{
                  "Name": "Jay",
                  "Age": "17"
              },
              "1":{
                  "Name": "Mark",
                  "Age": "15"
              },
              "2":{
                  "Name": "Jack",
                  "Age":"16"
              }
          }
    '''

df7 = pd.read_json(data, orient ='index')
```

```python
df7.head()
```
